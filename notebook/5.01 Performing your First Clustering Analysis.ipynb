```python
import pandas as pd
from sklearn.cluster import KMeans

file_url = r'https://raw.githubusercontent.com/sedeba19/Chapter-5/main/data/DataSet_taxstats2015.csv'
```

```python
df = pd.read_csv(file_url)
```

```python
df.head()
```

```python
df.shape
```

```python
# Instantiate the KMeans class
kmeans = KMeans(random_state=1234)
```

```python
# Create a subset from a dataset
X = df[['Average net tax',
       'Average total deductions']]
X.head()
```

```python
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
y_pred
```

```python
df['cluster'] = y_pred
df.head()
```

```python
df['cluster'].value_counts()
```

```python
df['cluster'].value_counts().index.sort_values()
```
